## 간단한 용어설명

- 유동비율 : 기업의 단기부채 상환능력을 평가하는 비율, 높을수록 재무구조가 건전
- 부채비율 : 높을수록 자기 자본보다 차입한 돈이 더 많은 것
- ROE : 자기자본을 활용해 1년간 얼마나 벌어들였는가를 나타내는 수익성 지표
- ROIC : 해당 기업의 영업활동 수익성 지표

# 유동비율 그래프

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(x=nexonVisual.index, y='유동비율', data=nexonVisual, ax=ax[0], color='blue') #넥슨 유동비율 그래프는 파랑색
sns.barplot(x=ncVisual.index, y='유동비율', data=ncVisual, ax=ax[1], color='green') #엔씨 유동비율 그래프는 초록색
```

넥슨에 비해 상대적으로 nc가 꾸준히 성장세다.
역시 넥슨 비리사건이 크긴 하다

```python
plt.figure(figsize = (20, 5))

plt.plot(nexonVisual['유동비율'], marker='o', color='blue', label='넥슨 유동비율')
plt.legend(loc='upper right')

plt.plot(ncVisual['유동비율'], marker='o', color='green', label='엔씨 유동비율')
plt.legend(loc='upper right')

plt.show()
```

근데 이렇게 보니 엔씨가 성장세가 아니라 꾸준히 유지하는 거였다.

# 부채비율 그래프

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(x=nexonVisual.index, y='부채비율', data=nexonVisual, ax=ax[0], color='blue') #넥슨 부채비율 그래프는 파랑색
sns.barplot(x=ncVisual.index, y='부채비율', data=ncVisual, ax=ax[1], color='green') #엔씨 부채비율 그래프는 초록색
```

그림으로만 보면 `그 사건` 전에는 넥슨이 부채비율이 더 낮은 것 같은데..

```python
plt.figure(figsize = (20, 5))

plt.plot(nexonVisual['부채비율'], marker='o', color='blue', label='넥슨 부채비율')
plt.legend(loc='upper right')

plt.plot(ncVisual['부채비율'], marker='o', color='green', label='엔씨 부채비율')
plt.legend(loc='upper right')

plt.show()
```

다시보니 `그 사건` 전에도 비슷했다. 차라리 꾸준한 nc가 더 나은 듯?

# ROE 그래프

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(x=nexonVisual.index, y='ROE', data=nexonVisual, ax=ax[0], color='blue')
sns.barplot(x=ncVisual.index, y='ROE', data=ncVisual, ax=ax[1], color='green') 
```

그래도 nc는 ROE 지표가 0 이하로 떨어진 적은 없다.
나락 간 넥슨이랑 한번에 비교한다면

```python
plt.figure(figsize = (20, 5))

plt.plot(nexonVisual['ROE'], marker='o', color='blue', label='넥슨 ROE')
plt.legend(loc='upper right')

plt.plot(ncVisual['ROE'], marker='o', color='green', label='엔씨 ROE')
plt.legend(loc='upper right')

plt.show()
```

엎치락뒤치락 하다가 나락가버린 넥슨

## ROE 정규분포 비교

```python
f, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.distplot(nexonVisual["ROE"], color="blue", ax=axes[0])
sns.distplot(ncVisual["ROE"], color="green", ax=axes[1])
```

모양은 비슷한데, 범위가 좀 다르다.
- 넥슨은 중심 x값이 0에 가까운데, nc는 중심 x값이 15 근처다
- 잘 모르니까 어떻게 분석해야할지는 잘 모르겠다.
