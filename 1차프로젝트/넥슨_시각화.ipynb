# 수익성 지표(ROE, ROIC, 영업이익률)

## ROE(자기자본이익률)
- 자기자본이익률(ROE) = ( 당기순이익 / 자본총계 ) * 100 (%)
- 기업이 자기자본을 활용해 1년간 얼마를 벌어들였는가를 나타내는 대표적인 수익성 지표

```python
nexonVisual['ROE'].plot(kind='line', figsize=(20,5)) #ROE 그래프
```

```python
sns.distplot(nexonVisual['ROE'])
```

## ROIC(투하자본이익률)
- ( 세후순영업이익(NOPAT) / 영업투하자본(IC) ) * 100
  - 세후순영업이익(NOPAT) = 영업이익 - 법인세비용
  - 영업투하자본(IC) = ( 매출채권 + 재고자산 + 매입채무 ) + 유형자산 + 무형자산
- 영업활동을 위한 자산과 영업활동으로 인한 이익을 가지고 해당 기업의 영업활동 수익성을 나타내는 지표


```python
nexonVisual['ROIC'].plot(kind='line', figsize=(20,5)) #ROIC 그래프
```

```python
fig = plt.figure(figsize = (20, 5)) 
ax1 = fig.add_subplot()

ax1.plot(nexonVisual['ROE'], marker='o', color = 'black')
ax1.set_ylabel('ROE')
ax1.tick_params(axis='y', labelcolor='black')

ax2 = ax1.twinx()
ax2.plot(nexonVisual['ROIC'], marker='o', color = 'red')
ax2.set_ylabel('ROIC')
ax2.tick_params(axis='y', labelcolor='red')

# ax3 = ax1.twinx()
# ax3.plot(nexonVisual['영업이익률'], marker='o', color = 'blue')
# ax3.set_ylabel('영업이익률')
# ax3.tick_params(axis='y', labelcolor='blue')
# ax3.spines['right'].set_position(("axes", 1.05)) ## 오른쪽 옆에 y축 추가


plt.show() 
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(x=nexonVisual.index, y='ROE', data=nexonVisual, ax=ax[0], color='blue')
sns.barplot(x=nexonVisual.index, y='ROIC', data=nexonVisual, ax=ax[1], color='green')
#sns.barplot(x=nexonVisual.index, y='영업이익률', data=nexonVisual, ax=ax[2], color='black')
```
