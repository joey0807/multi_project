```python
col_list = ['temp'] #지표 테이블 만들어주기
for i in range(len(nexon.columns)):
    col_list.append(nexon.columns[i])   

col_list.append('temp')

nexonTable = pd.DataFrame(columns=col_list, index = ['유동비율', '부채비율', 'ROE', 'ROIC', '영업이익률', '매출액증가율', '자기자본증가율'])
```

```python
nexonTable
```

```python
nexonTable.drop(columns=['자산종류'], inplace=True) #필요없는 부분 컷
```

```python
nexonTable
```

```python
for i in range(1,len(rawData.columns)): #값 넣어주기
    nexonTable.loc['유동비율', rawData.columns[i]] = ((rawData.loc[0, rawData.columns[i]] / rawData.loc[3, rawData.columns[i]]) * 100)
    nexonTable.loc['부채비율', rawData.columns[i]] = ((rawData.loc[4,rawData.columns[i]] / rawData.loc[6, rawData.columns[i]]) * 100)
    nexonTable.loc['영업이익률', rawData.columns[i]] = ((rawData.loc[9, rawData.columns[i]] / rawData.loc[8, rawData.columns[i]]) * 100)
    nexonTable.loc['매출액증가율', rawData.columns[i]] = (((rawData.loc[8, rawData.columns[i]] - rawData.loc[8, rawData.columns[i-1]]) / rawData.loc[8, rawData.columns[i-1]]) * 100)
    nexonTable.loc['자기자본증가율', rawData.columns[i]] = (((rawData.loc[6, rawData.columns[i]] - rawData.loc[6, rawData.columns[i-1]]) / rawData.loc[6, rawData.columns[i-1]]) * 100)
    nexonTable.loc['ROE', rawData.columns[i]] = ((rawData.loc[10,rawData.columns[i]] / ((rawData.loc[6, rawData.columns[i-1]] + rawData.loc[6, rawData.columns[i]])/2)) * 100)
    nexonTable.loc['ROIC', rawData.columns[i]] = (((rawData.loc[9, rawData.columns[i]] - rawData.loc[11, rawData.columns[i]]) / ((rawData.loc[2, rawData.columns[i]] + rawData.loc[2, rawData.columns[i]])/2)) * 100)
```

```python
nexonTable
```

```python
nexonTable.drop(nexonTable.iloc[:, :2], axis=1, inplace=True)
```
